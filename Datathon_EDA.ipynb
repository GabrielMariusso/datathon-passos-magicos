```python
# Padronização da coluna 'FASE'

# 2022: 
df_2022['Fase'] = df_2022['Fase'].apply(lambda x: 'ALFA' if x == 0 else f'FASE {x}')

# 2024:
def padronizar_fase_2024(valor):
    valor = str(valor).strip()
    if valor == 'ALFA':
        return 'ALFA'
    numero = ''.join(filter(str.isdigit, valor))
    return f'FASE {numero}' if numero else valor

df_2024['Fase'] = df_2024['Fase'].apply(padronizar_fase_2024)

# Verificação
print("Fase 2022:", sorted(df_2022['Fase'].unique()))
print("Fase 2023:", sorted(df_2023['Fase'].unique()))
print("Fase 2024:", sorted(df_2024['Fase'].unique()))
```

```python
# Padronização da coluna 'Fase Ideal'

def padronizar_fase_ideal(valor):
    if pd.isna(valor):
        return valor
    valor = str(valor).strip()
    # Verificar se é ALFA
    if valor.upper().startswith('ALFA'):
        return 'ALFA'
    # Extrair só o número
    numero = ''.join(filter(str.isdigit, valor.split('(')[0]))
    return f'FASE {numero}' if numero else valor

df_2022['Fase Ideal'] = df_2022['Fase Ideal'].apply(padronizar_fase_ideal)
df_2023['Fase Ideal'] = df_2023['Fase Ideal'].apply(padronizar_fase_ideal)
df_2024['Fase Ideal'] = df_2024['Fase Ideal'].apply(padronizar_fase_ideal)

# Verificação
print("Fase Ideal 2022:", sorted(df_2022['Fase Ideal'].dropna().unique()))
print("Fase Ideal 2023:", sorted(df_2023['Fase Ideal'].dropna().unique()))
print("Fase Ideal 2024:", sorted(df_2024['Fase Ideal'].dropna().unique()))
```

```python
# Padronização da coluna 'Idade'

import datetime

def corrigir_idade(valor):
    if isinstance(valor, datetime.datetime):
        return valor.day  # extrai o dia, que é a idade real
    return int(valor)

df_2023['Idade'] = df_2023['Idade'].apply(corrigir_idade)

# Verificação
print("Tipo coluna Idade 2023:", df_2023['Idade'].dtype)
```

#### Efetividade do Programa

```python
# Criando colunas numéricas para todas as pedras
mapa_pedra = {
    'Quartzo':  1,
    'Agata':    2,
    'Ametista': 3,
    'Topazio':  4
}

for col in ['Pedra 20', 'Pedra 21', 'Pedra 22', 'Pedra 2023', 'Pedra 2024']:
    df_analise[col + '_num'] = df_analise[col].map(mapa_pedra)

# Verificando quantidade de dados disponíveis por ano
for col in ['Pedra 20', 'Pedra 21', 'Pedra 22', 'Pedra 2023', 'Pedra 2024']:
    print(f"{col}: {df_analise[col].notna().sum()} alunos")
```

```python
# Identificar alunos com dados em anos consecutivos
alunos_20 = set(df_analise[df_analise['Pedra 20'].notna()]['RA'])
alunos_21 = set(df_analise[df_analise['Pedra 21'].notna()]['RA'])
alunos_22 = set(df_analise[df_analise['Pedra 22'].notna()]['RA'])
alunos_23 = set(df_analise[df_analise['Pedra 2023'].notna()]['RA'])
alunos_24 = set(df_analise[df_analise['Pedra 2024'].notna()]['RA'])

print(f"Transição 2020 → 2021: {len(alunos_20 & alunos_21)} alunos")
print(f"Transição 2021 → 2022: {len(alunos_21 & alunos_22)} alunos")
print(f"Transição 2022 → 2023: {len(alunos_22 & alunos_23)} alunos")
print(f"Transição 2023 → 2024: {len(alunos_23 & alunos_24)} alunos")
```

```python
def classificar_evolucao(valor):
    if pd.isna(valor):
        return np.nan
    elif valor > 0:
        return 'Subiu'
    elif valor == 0:
        return 'Manteve'
    else:
        return 'Desceu'

def criar_transicao(df, alunos, col_inicio, col_fim, periodo):
    df_filtrado = df[df['RA'].isin(alunos)].groupby('RA').agg(
        Pedra_inicio = (col_inicio, 'first'),
        Pedra_fim    = (col_fim, 'first')
    ).reset_index()
    df_filtrado['Periodo'] = periodo
    return df_filtrado

trans_20_21 = criar_transicao(df_analise, alunos_20 & alunos_21, 'Pedra 20_num', 'Pedra 21_num', '2020 → 2021')
trans_21_22 = criar_transicao(df_analise, alunos_21 & alunos_22, 'Pedra 21_num', 'Pedra 22_num', '2021 → 2022')
trans_22_23 = criar_transicao(df_analise, alunos_22 & alunos_23, 'Pedra 22_num', 'Pedra 2023_num', '2022 → 2023')
trans_23_24 = criar_transicao(df_analise, alunos_23 & alunos_24, 'Pedra 2023_num', 'Pedra 2024_num', '2023 → 2024')

df_transicoes = pd.concat([trans_20_21, trans_21_22, trans_22_23, trans_23_24], ignore_index=True)

df_transicoes['Evolucao'] = df_transicoes['Pedra_fim'] - df_transicoes['Pedra_inicio']
df_transicoes['Status'] = df_transicoes['Evolucao'].apply(classificar_evolucao)

print(f"Shape: {df_transicoes.shape}")
print(f"\nEvolução por período:")
print(df_transicoes.groupby('Periodo')['Status'].value_counts())
```

```python
# Calcular percentual por período
df_status_pct = (df_transicoes.groupby('Periodo')['Status']
                 .value_counts(normalize=True)
                 .mul(100)
                 .round(1)
                 .reset_index())
df_status_pct.columns = ['Periodo', 'Status', 'Percentual']

print(df_status_pct)
```

```python
cores_status = {'Subiu': '#2ecc71', 'Manteve': '#f39c12', 'Desceu': '#e74c3c'}
ordem_status = ['Subiu', 'Manteve', 'Desceu']
periodos = ['2020 → 2021', '2021 → 2022', '2022 → 2023', '2023 → 2024']

fig, ax = plt.subplots(figsize=(12, 6))

bottom = np.zeros(len(periodos))
for status in ordem_status:
    valores = [df_status_pct[(df_status_pct['Periodo'] == p) & 
               (df_status_pct['Status'] == status)]['Percentual'].values[0] 
               for p in periodos]
    bars = ax.bar(periodos, valores, bottom=bottom, 
                  color=cores_status[status], label=status, edgecolor='white')
    for bar, val in zip(bars, valores):
        if val > 5:
            ax.text(bar.get_x() + bar.get_width()/2,
                    bar.get_y() + bar.get_height()/2,
                    f'{val:.1f}%', ha='center', va='center', 
                    fontsize=10, fontweight='bold', color='black')
    bottom += np.array(valores)

ax.set_title('Evolução da Classificação dos Alunos por Período', fontsize=13, fontweight='bold')
ax.set_xlabel('Período', fontsize=11)
ax.set_ylabel('Percentual (%)', fontsize=11)
ax.legend(fontsize=10, loc='upper left', bbox_to_anchor=(1, 1))
ax.set_ylim(0, 110)
plt.tight_layout()
plt.savefig('Evolução_alunos_por_periodo.png',dpi=300,bbox_inches='tight',facecolor='white')
plt.show()
```

```python
# Distribuição das pedras por ano
ordem_pedra = ['Quartzo', 'Agata', 'Ametista', 'Topazio']
cores_pedra = {'Quartzo': '#7fb3d3', 'Agata': '#82e0aa', 
               'Ametista': '#c39bd3', 'Topazio': '#f9e79f'}
anos = [2022, 2023, 2024]
cols_pedra = ['Pedra 22', 'Pedra 2023', 'Pedra 2024']

fig, ax = plt.subplots(figsize=(10, 6))

bottom = np.zeros(len(anos))
for pedra in ordem_pedra:
    valores = []
    for ano, col in zip(anos, cols_pedra):
        pct = (df_analise[df_analise['Ano'] == ano][col]
               .value_counts(normalize=True).mul(100).get(pedra, 0))
        valores.append(round(pct, 1))
    bars = ax.bar([str(a) for a in anos], valores, bottom=bottom,
                  color=cores_pedra[pedra], label=pedra, edgecolor='white')
    for bar, val in zip(bars, valores):
        if val > 3:
            ax.text(bar.get_x() + bar.get_width()/2,
                    bar.get_y() + bar.get_height()/2,
                    f'{val:.1f}%', ha='center', va='center',
                    fontsize=10, fontweight='bold', color='black')
    bottom += np.array(valores)

ax.set_title('Distribuição das Pedras por Ano', fontsize=13, fontweight='bold')
ax.set_xlabel('Ano', fontsize=11)
ax.set_ylabel('Percentual (%)', fontsize=11)
ax.legend(fontsize=10, loc='upper left', bbox_to_anchor=(1, 1))
ax.set_ylim(0, 110)
plt.tight_layout()
plt.savefig('Distribuição_pedras.png',dpi=300,bbox_inches='tight',facecolor='white')
plt.show()
```

Efetividade do Programa

A análise da evolução das pedras ao longo dos períodos mostra uma melhora gradual e consistente no programa. A taxa de alunos que subiram de nível cresceu de 14.9% (2020→2021) para 24.2% (2023→2024), enquanto a taxa de alunos que desceram de nível reduziu de 32.8% para 25.8% no mesmo período. Além disso, o número de alunos acompanhados cresceu significativamente, passando de 323 para 678 entre os períodos analisados, o que reforça tanto o alcance quanto o impacto positivo do programa Passos Mágicos ao longo dos anos.
A distribuição das pedras ao longo dos anos reforça o impacto positivo do programa. No mesmo período, as classificações mais baixas como Quartzo e Ágata apresentaram redução, indicando uma migração consistente dos alunos para níveis superiores de desenvolvimento.

```python
def categorizar_defasagem(valor):
    if valor >= 0:
        return 'Sem Defasagem'
    elif valor == -1:
        return 'Leve (-1)'
    elif valor == -2:
        return 'Moderada (-2)'
    else:
        return 'Severa (-3 ou menos)'

df_analise['Categoria_Defasagem'] = df_analise['Defasagem'].apply(categorizar_defasagem)

print("Distribuição geral:")
print(df_analise['Categoria_Defasagem'].value_counts())

print("\nDistribuição por ano (%):")
print(df_analise.groupby('Ano')['Categoria_Defasagem']
      .value_counts(normalize=True)
      .mul(100).round(1))
```

```python
ordem_cat = ['Sem Defasagem', 'Leve (-1)', 'Moderada (-2)', 'Severa (-3 ou menos)']
cores_cat = {
    'Sem Defasagem':        '#2ecc71',
    'Leve (-1)':            '#f39c12',
    'Moderada (-2)':        '#e67e22',
    'Severa (-3 ou menos)': '#e74c3c'
}

df_cat_pct = (df_analise.groupby('Ano')['Categoria_Defasagem']
              .value_counts(normalize=True)
              .mul(100).round(1)
              .reset_index())
df_cat_pct.columns = ['Ano', 'Categoria', 'Percentual']

anos = [2022, 2023, 2024]
fig, ax = plt.subplots(figsize=(10, 6))

bottom = np.zeros(len(anos))
for cat in ordem_cat:
    valores = [df_cat_pct[(df_cat_pct['Ano'] == a) & 
               (df_cat_pct['Categoria'] == cat)]['Percentual'].values[0] 
               for a in anos]
    bars = ax.bar([str(a) for a in anos], valores, bottom=bottom,
                  color=cores_cat[cat], label=cat, edgecolor='white')
    for bar, val in zip(bars, valores):
        if val > 3:
            ax.text(bar.get_x() + bar.get_width()/2,
                    bar.get_y() + bar.get_height()/2,
                    f'{val:.1f}%', ha='center', va='center',
                    fontsize=10, fontweight='bold', color='black')
    bottom += np.array(valores)

ax.set_title('Perfil de Defasagem dos Alunos por Ano', fontsize=13, fontweight='bold')
ax.set_xlabel('Ano', fontsize=11)
ax.set_ylabel('Percentual (%)', fontsize=11)
ax.legend(fontsize=10, loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.savefig('IAN.png',dpi=300,bbox_inches='tight',facecolor='white')
plt.show()
```

Adequação do Nível (IAN) e Perfil de Defasagem

A análise do perfil de defasagem dos alunos revela uma melhora expressiva ao longo dos anos. Em 2022, apenas 30.1% dos alunos estavam sem defasagem, enquanto em 2024 esse número subiu para 53.8%. A defasagem severa (-3 ou menos) praticamente desapareceu, caindo de 3.3% em 2022 para apenas 0.3% em 2024. A defasagem moderada também reduziu significativamente de 19% para 7.8% no mesmo período. No geral, 44.3% dos alunos estão sem defasagem e 41.6% apresentam defasagem leve, indicando que a maioria dos alunos está próxima ou dentro do nível esperado.
